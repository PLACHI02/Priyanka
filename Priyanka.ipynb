```python
class Car:
    # Class attribute (shared by all instances)
    max_speed = 120  # Maximum speed in km/h
    # Constructor method (initialize instance attributes)
    def __init__(self, make, model, color, speed=0):
        self.make = make
        self.model = model
        self.color = color
        self.speed = speed  # Initial speed is set to 0
    # Method for accelerating the car
    def accelerate(self, acceleration):
        if self.speed + acceleration <= Car.max_speed:
            self.speed += acceleration
        else:
            self.speed = Car.max_speed
    # Method to get the current speed of the car
    def get_speed(self):
        return self.speed
     # Create objects (instances) of the Car class
car1 = Car("Toyota", "Camry", "Blue")
car2 = Car("Honda", "Civic", "Red") #Accessing object attributes
# Accelerate the cars
car1.accelerate(30)
car2.accelerate(20)   #Modifying Object attributes
# Print the current speeds of the cars
print(f"{car1.make} {car1.model} is currently at {car1.get_speed()} km/h.")
print(f"{car2.make} {car2.model} is currently at {car2.get_speed()} km/h.")
```

```python
#Defining the base class  #Polymorphism Example
class Shape:
  def area(self):
   pass
#Defining the subclass - Recatngle
class Rectangle(Shape):
  def __init__(self,width,height):
   self.width = width
   self.height = height
  def area(self):
    return self.width * self.height
Shapes = [Rectangle(8,6)]#creating list i.e objects
for Shape in Shapes:
 print(f"Area:{Shape.area()}")


```

```python
class Shape:
  def area(self):
    pass
class Circle(Shape):
  def __init__(self,radius):
    self.radius = radius
  def area(self):
    return 3.14 * self.radius * self.radius
Shapes = [Circle(4)]
for Shape in Shapes:
  print(f"{Shape.area()}")    #polymorhism

```

- API in python allows you to interact with software services & hardware by providing a predefined protocols & method.

Eg - 1.Interacting with external web serivices(Twitter API & Google Maps API)

     2. Accessing system level operations (OS API's , File operations)
      
     3. Facilitating communication between components in an application

**TYPES OF API**
- **WEB API's** - These are HTTP based services which interact with web servers (RESTful API's, GraphQL API's and SOAP API's)
- **LIBRARY API's** - Functions & Classes exposed by libraries to be consumed in pythonscripts (Numpy & Pandas)
-**HARDWARE API's** - Enable the communication with hardware components like devices & sensors (GPIO API's & Raspberry Pi API's)
-**SYSTEM API's** - API's for interacting with operating systems (os module in python).

REST API's

```python
URL = "https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"

# Extract tables from webpage using Pandas. Retain table number 3 as the required dataframe.
tables = pd.read_html(URL)
df = tables[3]

# Replace the column headers with column numbers
df.columns = range(df.shape[1]) #df.shape returns to the tuple rep dimensions of dataframe (no:of Rows,no:of columns).df.shape[1] extracts column no

# Retain columns with index 0 and 2 (name of country and value of GDP quoted by IMF)
#the code is indexed rather than  using column names becoz in the above code we have changed column names to numbers
df = df[[0,2]] #[[]] indicates that result remains as dataframe (not series) and this operation doestnt modift rather it creates a new dataframe df with only selected columns 0 and 2

# Retain the Rows with index 1 to 10, indicating the top 10 economies of the world.
df = df.iloc[1:11,:] #iloc is used for index based selection in pandas

#Assign column names as "Country" and "GDP (Million USD)"
df.columns = ["Country", "GDP (Million USD)"]

# Change the data type of the 'GDP (Million USD)' column to integer. Use astype() method.
df["GDP (Million USD)"] = df["GDP (Million USD)"].astype(int)

# Convert the GDP value in Million USD to Billion USD
df[["GDP (Million USD)"]] = df[["GDP (Million USD)"]]/1000

# Use numpy.round() method to round the value to 2 decimal places.
df[["GDP (Million USD)"]] = np.round(df[["GDP (Million USD)"]],2)


# Rename the column header from 'GDP (Million USD)' to 'GDP (Billion USD)'
df.rename(columns = {"GDP (Million USD)" : "GDP (Billion USD)"})

#Loading dataframe to csv file named Largest_economies.csv
df.to_csv("./Largest_economies.csv")

```

**REST API AND HTTP REQUEST**
-HTTP protocol is a general protocol of transferring information through web.
-REST API function by sending a request & is communicated via HTTP message.The HTTP message usually contains JSON file.
- When the clinet use a webpage,their browser sends an HTTP request to the server where th webpage is hosted.
-The server tries to find the desired resource by default **index.html**.
-If your request is succesful, the server send the send the object to client in an **HTTP Response**, this includes info related to type of resource,lenght of resource and other info.

-When a request is made for information,the webservers send the information in txt.file or png.file or html.file.

-URL -**Uniform Resource locator** is the most popular way to find resources on web. which has 3 main parts

   Scheme : this is the protocol & eg : **http://**

  Internet Address or Base URL - used to find Location of webserver & Eg :** www.ibm.com**

  Route - location on the webserver

-**REQUEST & RESPONSE**  
- Request Start Line - **/Get/Index.html HTTP/1.0** (This is Get method or HTTP method which is requesting the file Index.Html).
-Request Header - which passes additional information with an HTTP request in Get method Request Header is empty.
-Request Body - Some request have a body


-Response Start Line - HTTP/1.0 200 OK (contains version number HTTP/1.0 followed by a status code 200 meaning success with descriptive phase OK).
-Response Header - contains information.
-Response Body - contains requested file, in this case HTML.


-Some examples of Status codes - 100's - Informational Responses, 200's - Successful Responses, 300's - Redirection Responses, 400's - Clinet Errors, 500'S - Server Errors

```python
import json
person = {"First Name":"Priyanka","Last name":"Lachchi","Age":28,"Address":{"Door NO":8,"Street Name":"Wayfaring Road","City":"Northampton","Pin Code":"NN4 6FL"}}

with open("person.json","w") as f:
  json.dump(person,f)
  print(person)
```

**WORKING WITH HTML TABLE**

-These are the following tags which are used while creating HTML tables

- **table**: This tag is the root tag used to define the start and end of the table. all the contents ofv table are enclosed within this tag.
- **tr**: This tag is used for table row.Each row of table is defined within this tag.
- **td**: This tag is used for table cell data. Each cell of table is defined within this tag.
- **th** : This tag is used for table header cell.This header cell is used to decribe the contents of a column or row
- **tbody**:This is the main content of the table



```python
#Step 4 Use BeautifulSoup method for extracting data
for row in soup.find("tbody").find_all("tr"): #First islolating the body of the table which contains all info & then we loop through each row and find all the column values for each row
    col = row.find_all("td") #Inside the tr find the table cell data. Each td represents a cell in the row
    date = col[0].text
    Open = col[1].text
    high = col[2].text
    low = col[3].text
    close = col[4].text
    adj_close = col[5].text
    volume = col[6].text


#Finally we append the data of each row to table
netflix_data = pd.concat([netflix_data,pd.DataFrame({"Date":[date],"Open":[Open],"High":[high],"Low":[low],"Close":[close],"Adj Close":[adj_close],"Volume":[volume]})],ignore_index=True)

```
